```
# File location and type
file_location = "/FileStore/tables/Premier_League_Player_Statistics_Final_Dataset.csv"
file_type = "csv"

# CSV options
infer_schema = "true"
first_row_is_header = "true"
delimiter = ";"

# The applied options are for CSV files. For other file types, these will be ignored.
epl_data = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .option("encoding", "ISO-8859-1") \
  .load(file_location)

display(epl_data)
```

```
# Dutch players dataframe
dutch_players = epl_data.filter(epl_data['Nation'] == 'NED')

# Filter young players (<21)
young_players = epl_data.filter(epl_data['Age'] < 21)

# Filter forwards
forwards = epl_data.filter(epl_data['Pos'].isin(['FW', 'FW,MF', 'MF,FW', 'FW,DF', 'DF,FW']))

# Displaying to test if it works
display(dutch_players)
```

```
from pyspark.sql.functions import col, sqrt

# Calculate the Euclidean distance for each forward based on relevant statistics
forwards_with_similarity = forwards.withColumn(
    'Similarity',
    sqrt(
        (col('Goals') - 19) ** 2 +
        (col('Assist') - 4) ** 2 +
        (col('xG') - 17.5) ** 2 +
        (col('NoPen_XG') - 15.1) ** 2 +
        (col('xAG') - 3.0) ** 2
    )
)

# Rank the forwards based on similarity
replacement_candidates = forwards_with_similarity.orderBy('Similarity')

# Select only the desired columns
replacement_candidates_select = replacement_candidates.select('Player', 'Nation', 'Pos', 'Squad', 'Age', 'Goals', 'Assist', 'xG', 'NoPen_XG', 'xAG')

# Display the top 5 replacement candidates as a table
display(replacement_candidates_select.limit(5))
```

```
from pyspark.sql.functions import col, rank
from pyspark.sql.window import Window

# Rank Dutch players based on performance metrics
dutch_players = dutch_players.withColumn(
    'Performance_Score',
    (col('Match_play') * 0.1) +
    ((col('Goals') + col('Assist')) * 0.2) +
    ((col('Prog_Car') + col('Prog_Pas')) * 0.1) +
    ((col('Tackle_Intercept') + col('Ball_Blocked')) * 0.2) +
    ((col('TotalComp_Passes') * col('TotalPas_Rate')) * 0.4)
)

# Rank Dutch players based on age
dutch_players = dutch_players.withColumn(
    'Age_Rank',
    rank().over(Window.orderBy(col('Age').desc()))
)

# Combine the ranks using a weighted average or any other method you prefer
dutch_players = dutch_players.withColumn(
    'Overall_Rank', 
    col('Performance_Score') * 0.6 +
    col('Age_Rank') * 0.4
)

# Sort players by overall rank to find the most interesting ones for Feyenoord
most_interesting_dutch_players = dutch_players.orderBy(col('Overall_Rank').desc()).limit(10)

# Select only the desired columns
most_interesting_dutch_players_select = most_interesting_dutch_players.select('Player', 'Pos', 'Squad', 'Age')

display(most_interesting_dutch_players_select)
```

```
# Group the DataFrame by age and count the number of players in each age group
age_distribution = young_players.groupby('Age').count()

# Sort the DataFrame by the 'Age' column in ascending order
age_distribution = age_distribution.orderBy(col('Age'))

# Display the table
age_distribution.display()
```

```
from pyspark.sql.functions import avg, format_number

# Group the DataFrame by age and calculate the average of matches played for each age group
matches_vs_age_avg = young_players.groupBy('Age').avg('Match_Play')

# Rename the avg column to 'Avg_Matches_Played'
matches_vs_age_avg = matches_vs_age_avg.withColumnRenamed('avg(Match_Play)', 'Avg_Matches_Played')

# Format the average matches played to display whole numbers
matches_vs_age_avg = matches_vs_age_avg.withColumn('Avg_Matches_Played', format_number('Avg_Matches_Played', 0))

# Sort the DataFrame by the 'Age' column in ascending order
matches_vs_age_avg = matches_vs_age_avg.orderBy('Age')

# Display the result
matches_vs_age_avg.display()
```

```
# Select the columns 'Player', 'Age', and 'Match_Play' from the DataFrame
matches_per_player = young_players.select('Player', 'Age', 'Match_Play')

# Sort the DataFrame by the 'Match_Play' column in descending order
matches_per_player = matches_per_player.orderBy(-matches_per_player['Match_Play'])

# Display the result
matches_per_player.display()
```
